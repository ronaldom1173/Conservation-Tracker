````python
import sys
sys.path.append('..')  # so we can import from src/

import os
import pandas as pd
import plotly.express as px

from src.iucn import get_species_page, get_species_history, get_species_by_category
from src.db import init_db, df_to_table, query
````

````python
# Set up the database (creates tables if they don't exist)
init_db()
````

## 1. Pull all species (paged)

The IUCN API returns ~10,000 species per page. We loop through all pages until we get everything.

````python
all_species = []
page = 0

while True:
    print(f"Fetching page {page}...")
    result = get_species_page(page=page)
    batch = result.get("result", [])
    if not batch:
        break
    all_species.extend(batch)
    page += 1

df_species = pd.DataFrame(all_species)
print(f"\nTotal species fetched: {len(df_species):,}")
df_species.head()
````

````python
# Save raw data to CSV
df_species.to_csv('../data/raw/iucn_species_all.csv', index=False)
print("Saved to data/raw/iucn_species_all.csv")
````

````python
# Save to database
df_to_table(df_species, 'species', if_exists='replace')
print("Saved to database table: species")
````

## 2. Breakdown by Red List category

Categories from most to least threatened:
- **EX** — Extinct
- **EW** — Extinct in the Wild
- **CR** — Critically Endangered
- **EN** — Endangered
- **VU** — Vulnerable
- **NT** — Near Threatened
- **LC** — Least Concern
- **DD** — Data Deficient

````python
category_order = ['EX', 'EW', 'CR', 'EN', 'VU', 'NT', 'LC', 'DD']

category_counts = (
    df_species['category']
    .value_counts()
    .reindex(category_order, fill_value=0)
    .reset_index()
)
category_counts.columns = ['category', 'count']

fig = px.bar(
    category_counts,
    x='category',
    y='count',
    color='category',
    title='Species Count by Red List Category',
    labels={'count': 'Number of Species', 'category': 'Red List Category'}
)
fig.show()
````

## 3. Breakdown by taxonomic group (class)

Which animal groups — mammals, birds, amphibians, etc. — have the most threatened species?

````python
threatened = ['CR', 'EN', 'VU']

df_threatened = df_species[df_species['category'].isin(threatened)]

by_class = (
    df_threatened
    .groupby('class_name')['taxonid']
    .count()
    .sort_values(ascending=False)
    .head(15)
    .reset_index()
)
by_class.columns = ['class', 'threatened_species']

fig2 = px.bar(
    by_class,
    x='threatened_species',
    y='class',
    orientation='h',
    title='Top 15 Classes by Number of Threatened Species (CR + EN + VU)',
    labels={'threatened_species': 'Threatened Species Count', 'class': 'Class'}
)
fig2.show()
````

## 4. Historical assessments — success stories

Pull the full assessment history for a few well-known species that improved status.

````python
# Examples of species that improved — feel free to add more
success_species = [
    "Haliaeetus leucocephalus",  # Bald Eagle
    "Megaptera novaeangliae",    # Humpback Whale
    "Panthera tigris",           # Tiger (partial recovery in some regions)
]

history_rows = []

for name in success_species:
    print(f"Fetching history for {name}...")
    result = get_species_history(name)
    for entry in result.get('result', []):
        entry['query_name'] = name
        history_rows.append(entry)

df_history = pd.DataFrame(history_rows)
df_history.head(20)
````

````python
# Map category codes to numeric threat level for plotting
threat_level = {'LC': 1, 'NT': 2, 'VU': 3, 'EN': 4, 'CR': 5, 'EW': 6, 'EX': 7}

df_history['threat_score'] = df_history['category'].map(threat_level)
df_history['year'] = pd.to_datetime(df_history['assessment_date']).dt.year

fig3 = px.line(
    df_history,
    x='year',
    y='threat_score',
    color='query_name',
    markers=True,
    title='Red List Status Over Time — Success Stories',
    labels={'threat_score': 'Threat Level (1=LC → 7=EX)', 'year': 'Year', 'query_name': 'Species'}
)
fig3.update_yaxes(tickvals=list(threat_level.values()), ticktext=list(threat_level.keys()))
fig3.show()
````

````python
# Save history to DB
df_to_table(df_history, 'assessments', if_exists='replace')
print("Saved to database table: assessments")
````
